```python
def outliers(data,col):
    Q1=data[col].quantile(0.25) 
    Q3=data[col].quantile(0.75) 
    IQR=Q3-Q1 
    max_value=Q3+IQR*1.5
    min_value=Q1-IQR*1.5
    return data[(data[col]<=min_value) | (data[col]>=max_value)]
print(outliers(data,'Maxpulse').shape)
```
